```python
import os
VIDEO_PATH = "C:\\Users\\Marco\\Desktop\\videos05\\videos"
ANNOTATIONS = "Annot_List.txt"
```

```python
# Get all filenames in the video folder
video_files = os.listdir(VIDEO_PATH)
video_files = [f for f in video_files if f.endswith(".avi")]
# Remove the .avi extension
video_files = [f[:-4] for f in video_files]
video_files
```

```python
class Annotation:
    def __init__(self, line):
        parts = line.strip().split(",")
        self.video = parts[0]
        self.label = parts[1]
        self.id = int(parts[2])
        self.start = float(parts[3])
        self.end = float(parts[4])
        self.duration = self.end - self.start
```

```python
# Open the annotations file and search for the videos
annotations = []

with open(ANNOTATIONS, "r") as f:
    lines = f.readlines()
    for line in lines:
        line_edit = line.strip().split(",")[0]
        if line_edit in video_files:
            print("Found: ", line)
            annotations.append(Annotation(line))
```

```python
from moviepy.video.io.VideoFileClip import VideoFileClip

# Funzione per croppare un video
def crop_video(input_path, output_path, start_frame, end_frame, fps=30):
    """
    Croppa un video dato il frame di inizio e fine.
    
    Args:
        input_path (str): Path del video originale.
        output_path (str): Path del video croppato in uscita.
        start_frame (int): Frame di inizio.
        end_frame (int): Frame di fine.
        fps (int): Frame per secondo del video originale (default: 30).
    """
    # Calcola il tempo di inizio e fine in secondi
    start_time = start_frame / fps
    end_time = end_frame / fps
    
    # Carica il video
    clip = VideoFileClip(input_path)
    
    # Taglia il video
    cropped_clip = clip.subclipped(start_time, end_time)
    
    # Salva il video tagliato
    cropped_clip.write_videofile(output_path, codec="libx264", audio_codec="aac")
    print(f"Video salvato in {output_path}.")

## Esempio di utilizzo
#input_video = "input_video.mp4"
#output_video = "cropped_video.mp4"
#start_frame = 300  # Inizio (frame)
#end_frame = 900    # Fine (frame)
#fps = 30           # Frame per secondo del video
#
#crop_video(input_video, output_video, start_frame, end_frame, fps)
```

```python
OUTPUT_DIR = "C:\\Users\\Marco\\Desktop\\videos05\\videos_cropped"
```

```python
for a in annotations:
    print(a.video, a.label, a.id, a.start, a.end, a.duration)
    if not os.path.exists(os.path.join(OUTPUT_DIR, a.label)):
        os.makedirs(os.path.join(OUTPUT_DIR, a.label))
    crop_video(os.path.join(VIDEO_PATH, a.video + ".avi"), os.path.join(OUTPUT_DIR, a.label, a.video + str(a.id) +".avi"), a.start, a.end)
    
```
